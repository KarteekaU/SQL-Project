## Visualization 22
#### The head of Slytherin wants to find all the Blood Status of his member students to put in his final house diversity report.

```python
result = student.merge(house, left_on='houseID', right_on='houseID')
Slytherin = result[result["houseName"] == 'Slytherin']
Slytherin['Blood status'].value_counts().plot(kind='bar', color = 'red')

font1 = {'family':'serif','color':'black','size':45}
font2 = {'family':'serif','color':'white','size':30}
plt.title("Visualization 2", fontdict = font1)
plt.xlabel("Blood Status", fontdict = font2)
plt.xticks(fontsize=25, color = 'white', rotation = 45)
plt.ylabel("Number of Students", fontdict = font2)
plt.yticks(fontsize=25, color = 'white')
```
